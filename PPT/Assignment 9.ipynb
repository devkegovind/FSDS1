Given an integerÂ `n`, returnÂ *`true`Â if it is a power of two. Otherwise, returnÂ `false`*.

An integerÂ `n`Â is a power of two, if there exists an integerÂ `x`Â such thatÂ `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

```python
def isPowerOfTwo(n):
    if n <= 0:
        return False
    while n % 2 == 0:
        n /= 2
    return n == 1
```

```python
print(isPowerOfTwo(1))  
print(isPowerOfTwo(16))  
print(isPowerOfTwo(3))  
```

Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

```python
def factorial(N):
    result = 1
    for i in range(1, N + 1):
        result *= i
    return result
```

```python
print(factorial(5))  
print(factorial(4))  
```

ðŸ’¡ **Question 5**

Given an array of integersÂ **arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

```python
def findMax(arr):
    # Base case: if the array contains only one element, return that element
    if len(arr) == 1:
        return arr[0]
    
    # Divide the array into two halves
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Recursively find the maximum elements of the left and right subarrays
    max_left = findMax(left)
    max_right = findMax(right)
    
    # Compare the maximum elements of the left and right subarrays and return the maximum
    return max(max_left, max_right)
```

```python
arr1 = [1, 4, 3, -5, -4, 8, 6]
print(findMax(arr1))  

arr2 = [1, 4, 45, 6, 10, -8]
print(findMax(arr2))  
```

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23


```python
def findNthTerm(a, d, N):
    nthTerm = a + (N - 1) * d
    return nthTerm
```

```python
print(findNthTerm(2, 1, 5))  
print(findNthTerm(5, 2, 10))  
```

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = â€œABCâ€*

***Output:***

*â€œABCâ€, â€œACBâ€, â€œBACâ€, â€œBCAâ€, â€œCBAâ€, â€œCABâ€*

**Example 2:**

***Input:***

*S = â€œXYâ€*

***Output:***

*â€œXYâ€, â€œYXâ€*

```python
def permute(s, l, r):
    if l == r:
        print("".join(s))
    else:
        for i in range(l, r + 1):
            # Fix character at position l
            s[l], s[i] = s[i], s[l]
            
            # Recursively generate permutations for the remaining characters
            permute(s, l + 1, r)
            
            # Backtrack and restore the original order
            s[l], s[i] = s[i], s[l]

def printPermutations(S):
    s = list(S)
    n = len(s)
    permute(s, 0, n - 1)
```

```python
printPermutations("ABC")

printPermutations("XY")

```

ðŸ’¡ **Question 8**

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18


```python
def productOfArray(arr):
    product = 1
    for num in arr:
        product *= num
    return product
```

```python
arr1 = [1, 2, 3, 4, 5]
print(productOfArray(arr1))  

arr2 = [1, 6, 3]
print(productOfArray(arr2)) 
```

**Bagging:**

In machine learning, bagging (short for bootstrap aggregating) is a technique that involves training multiple models on different subsets of the training data and then combining their predictions to make a final prediction. Bagging is commonly used to improve the performance and robustness of machine learning algorithms, especially decision tree-based models.

Multiple Models runs parallely.

It takes only Homogenous Models(similar models)

Here's how the bagging process typically works:

1. **Bootstrap Sampling**: Given a training dataset with N samples, bagging randomly selects a subset of the data of size N (usually with replacement). This means that some samples may appear multiple times in the subset, while others may be left out.

2. **Model Training**: For each subset, a base model (e.g., decision tree, neural network, etc.) is trained independently on the selected subset. This process is repeated M times, where M is the number of models or iterations.

3. **Prediction Aggregation**: Once the models are trained, they are used to make predictions on unseen data. For classification tasks, the most common aggregation method is voting, where each model's prediction is treated as a vote, and the majority vote is considered as the final prediction. For regression tasks, the predictions can be averaged to obtain the final prediction.

Bagging helps reduce the variance in the model's predictions by introducing randomness through bootstrapping and combining diverse models. It can also handle overfitting, improve model generalization, and provide better stability when faced with noisy or incomplete data.

Random Forest is a popular ensemble model that uses bagging by combining multiple decision trees trained on different subsets of the data. Each decision tree is trained independently, and their predictions are aggregated to make the final prediction.

Bagging can be applied to various machine learning algorithms, not just decision trees. The idea is to create an ensemble of diverse models to improve overall performance and robustness.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
models = {
    "RandomForest" : RandomForestClassifier(),
    "LogisticRegression" : LogisticRegression(),
    "DecisionTreeClassifier" : DecisionTreeClassifier()
}

def evaluate_model(X_train, X_test, y_train, y_test, models):
    report = {}
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = round(accuracy_score(y_test, y_pred) * 100, 2)
        report[model_name] = acc
    return report
```

```python
evaluate_model(X14_train, X14_test, y13_train, y13_test, models)
```

**Boosting:**

Boosting is another ensemble learning technique in machine learning that combines multiple weak or base models to create a stronger predictive model. Unlike bagging, which trains models independently, boosting builds models sequentially, where each subsequent model focuses on correcting the mistakes made by the previous models. The main idea behind boosting is to give more weight to the misclassified instances in order to improve the overall accuracy of the model.

Multiple Models runs sequently.

It also takes only Homogenous Models(similar models)

Here's a general overview of the boosting process:

1. **Base Model Training**: Initially, a base or weak model is trained on the entire training dataset. The weak model could be a simple decision tree, often called a decision stump, which is a tree with only one split.

2. **Instance Weighting**: Each instance in the training set is assigned an initial weight. Initially, all weights are usually set to equal values, meaning each instance has the same importance.

3. **Model Iterations**: The boosting algorithm then proceeds through a series of iterations. In each iteration, a new weak model is trained, with a particular focus on the instances that were misclassified by the previous models. The instances that were misclassified are assigned higher weights, making them more influential in the subsequent model training.

4. **Model Combination**: As the iterations continue, the weak models are combined to create a stronger ensemble model. The combination process typically involves assigning weights to each weak model based on its performance. Models that perform better in terms of classification accuracy are given higher weights in the final model.

5. **Prediction**: Finally, when making predictions on unseen data, the boosted model combines the predictions of all the weak models, with each model's prediction weighted by its assigned weight. The combined predictions provide the final prediction of the boosting model.

Popular boosting algorithms include AdaBoost (Adaptive Boosting) and Gradient Boosting. AdaBoost places higher weights on misclassified instances to improve their classification in subsequent iterations. Gradient Boosting, on the other hand, uses the gradient descent optimization technique to iteratively train weak models that minimize the loss function.

Boosting is effective in handling complex datasets and achieving high predictive accuracy. It can often outperform individual base models and has been successful in various domains, such as image classification, text classification, and regression tasks.

**Stacking:**

In machine learning, stacking (short for stacked generalization) is an ensemble learning technique that combines multiple models by training a meta-model to make predictions based on the outputs of the individual models. Stacking goes beyond simple averaging or voting, as it learns how to best combine the predictions of the base models to make a final prediction.

Multiple Model in two stages.

It take Heterogenous Models(Different Models)

Here's an overview of the stacking process:

1. **Base Model Training**: Initially, multiple base models (e.g., decision trees, support vector machines, neural networks, etc.) are trained on the training dataset. Each base model learns to make predictions based on the input features and produces its own set of predictions.

2. **Holdout Set Creation**: A holdout set is created from the training data that is not used during the base model training. This holdout set will be used to train the meta-model in the next step.

3. **Meta-Model Training**: The holdout set is then used to train a meta-model or a blender model. The holdout set contains the predictions made by the base models as the input features, and the target values of the training dataset as the target variable. The meta-model learns how to combine or weigh the predictions of the base models to make the final prediction.

4. **Prediction**: Once the meta-model is trained, it can be used to make predictions on new, unseen data. The base models make predictions on the input features, and these predictions are then fed into the meta-model, which produces the final prediction.

Stacking leverages the strengths of the individual base models and aims to improve the overall predictive performance by learning to combine their outputs effectively. It allows for more complex interactions and can capture patterns that might be missed by individual models.

Stacking can be used in various machine learning tasks such as classification, regression, or even unsupervised learning. It provides flexibility in choosing diverse base models and the ability to incorporate domain knowledge into the meta-model's training process.

However, it's important to note that stacking requires careful implementation and can be computationally expensive, as it involves training multiple models and a meta-model. Additionally, overfitting can be a concern if not properly controlled, so techniques like cross-validation and regularization are often used to mitigate this risk.

Overall, stacking is a powerful technique for combining the predictions of multiple models and has been successful in various machine learning competitions and real-world applications.
